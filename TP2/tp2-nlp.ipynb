```python
from num2words import num2words
import re

def convert_numbers_to_words(text):
    
    words = text.split()
    new_words = []
    
    for w in words:
        if w.isdigit():
            new_words.append(num2words(int(w)))
        else:
            new_words.append(w)
    
    return " ".join(new_words)
```

![ ](https://img.shields.io/badge/drop_stop_words%20%20-blue)

```python
stop_words = ["the","a","an","and","is","are","in","on","at","to","for","of","with"]

def drop_stop_words(text):
    words = text.split()
    words = [w for w in words if w not in stop_words]
    return " ".join(words)
```

![ ](https://img.shields.io/badge/drop_punctuations%20%20-blue)

```python
import string

def drop_punctuations(text):
    for p in string.punctuation:
        text = text.replace(p, "")
    return text
```

![ ](https://img.shields.io/badge/Process%20%20-orange)

```python
def process(text):

    text = text.lower()

    text = convert_numbers_to_words(text)

    text = drop_punctuations(text)

    text = drop_stop_words(text)

    return text


processed_X = []
for doc in X:
    processed_X.append(process(doc))
```

![ ](https://img.shields.io/badge/Vocabs%20%20-blue)

```python
vocab = []

for doc in processed_X:
    words = doc.split()
    for w in words:
        if w not in vocab:
            vocab.append(w)

print("Taille vocabulaire:", len(vocab))
```

## BOW
>1-gram BOW

[![Naive Bayes](https://img.shields.io/badge/Naive_Bayes-classification-pink)](https://digitalpro.dev)

> Convert X docs to NumPy array

```python
Vectors = []

for doc in processed_X:

    words = doc.split()
    vector = []

    for v in vocab:
        if v in words:
            vector.append(1)
        else:
            vector.append(0)

    Vectors.append(vector)
```

```python
import numpy as np

X_vectors = np.array(Vectors) #define your Vectors
y_array = np.array(y)
```

> Split into train/test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_vectors, y_array, test_size=0.2, random_state=42, stratify=y_array
)
```

> Train Naive Bayes classifier

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score, classification_report

nb = BernoulliNB()
nb.fit(X_train, y_train)    # Train

y_pred = nb.predict(X_test) # Predict
```

> Evaluate

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
